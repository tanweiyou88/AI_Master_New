        2) You will start with extracting/detecting the tiny features (EG: little edges), then you go to extract more general features (EG: eye, nose, ears). Then, you go to extract more more general features (EG:head and body), and then you do flattening again.
        3) The main idea behind convolutional neural network is feature extraction because the second part is a simple artificial neural network. But by doing this convolution, you are detecting the features & reducing the dimension
        4) You will specify a hyperparameter of how many filters you want to have in your CNN and what is the size of each of the filters. That's it but you do not specify the exact values (pixel weightages) within these filters, the network will learn those on its own.
