```python
import torch

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

"""
y = torch.tensor([1,1,1,1])
y = y.to(device)

z = y + y

print(z)
"""
# References:
# 1) https://stackoverflow.com/questions/64111559/how-to-understand-tensors-with-multiple-dimensions
# 2) https://www.youtube.com/watch?v=JXfDlgrfOBY
#
# a scalar = a Rank 0 tensor (EG: [1]; having shape of (1) )
# a vector = a Rank 1 tensor (EG: [1, 2, 3]; having shape of (1,3) )  *(row = 1st dimension, column = 2nd dimension)
# a matrix = a Rank 2 tensor (EG: [[1, 2, 3],[1, 2, 3]]; having shape of (2,3) ) *(row, column)
# a 3D tensor = a Rank 3 tensor (EG: [[[1, 2, 3],[1, 2, 3]],[[1, 2, 3],[1, 2, 3]]]; having shape of (2,2,3) ) *(channel = 1st dimension, row = 2nd dimension, column = 3rd dimension)

x3 = torch.ones(32,3,2) # x3 is a 3D tensor = a list of matrices = a list of lists of rows (channel = 1st dimension, row = 2nd dimension, column = 3rd dimension)
x3 = x3.to(device)
print(x3)

x4 = torch.ones(32,3,2) * 2 # (channel, row, column)
x4 = x4.to(device)
print(x4)

b = torch.cat([x3, x4],1) # (channel, row, column)
print(b)
```
